```python
# import library
from pytwmap import *
from IPython.display import display  # type: ignore
```

```python
# load a map from file
m = TWMap()
m.open('data/HeyTux2.map')
```

```python
# show all the groups of this map (in order as they would be in teeworlds)
# the hex numbers indicate that the group is unnamed
m.groups
```

```python
# change the name of one group
m.groups[0].name = 'quads_0'
m.groups
```

```python
# list the layers of each group
for g in m.groups:
    print(g.layers)
```

```python
# this is just a shortcut for the above cell
m.layers
```

```python
# shortcuts to get all non-design layers
# m.design_layers -> get all desing layers
m.gameplay_layers
```

```python
# access special ddnet layers
m.tele_layer
```

```python
# properties of layers:
if m.tele_layer:  # check if there actually is a tele layer
    print(m.tele_layer.width, m.tele_layer.height)
```

```python
# get the image of a design-layer
layer = m.layers[9]
if isinstance(layer, ItemTileLayer):  # could also be quad/sound layer
    if layer.image:  # could also be None
        display(layer.image.image)
```

```python
# save the map
m.save('data/HeyTux2_saved.map')
```
