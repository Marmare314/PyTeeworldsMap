```python
# import library
from pytwmap import *
```

```python
# load a map from file
m = TWMap()
m.open('data/HeyTux2.map')
```

```python
# create a renderer which outputs images of size 1000x1000
r = MapRenderer(map_ref=m, image_size=(1000, 1000))

# generate the tilesets, if images are changed/added this has to be done again
r.generate_tilesets()
```

```python
# clear the internal image buffer
# otherwise if this cell is executed twice the old image may still show in some places
r.clear_buffer()

# render the map in design
# x_start, y_start: the ingame coordinates of the top-left corner
# tile_scale: size of each tile in pixels
r.render_map_design(100 * 32, 0, tile_scale=10)
```

```python
# get image buffer as a pillow image so it can be displayed here
r.get_image()
```

```python
r.clear_buffer()
r.render_map_gameplay(100 * 32, 0, tile_scale=10)
r.get_image()
```
